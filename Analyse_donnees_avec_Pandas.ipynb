```python
# Import pandas
import pandas as pd
import numpy as np
```

```python
# Lire le fichier cars.csv, et assigné le résultat à la variable data
data = pd.read_csv('cars.csv')
```

```python
# Le nombre de lignes et de colonnes
print(data.shape)
print(f'{data.shape[0]} lignes et {data.shape[1]} colonnes.')
```

```python
# Liste des noms des colonnes
print(f'La liste des noms des colonnes :\n{list(data.columns)}')
```

```python
# Informations sur les colonnes:
data.info()
```

```python
# Voir les cinq première lignes
data.head()
```

```python
# Voir les cinq dérnières lignes
data.tail()
```

```python
# Le nombre des valeurs manquantes par colonnes
# On voit qu'on a 6 valeurs manquantes dans la colonnes horsepower
data.isnull().sum()
```

```python
# Affiches les statistiques des colonnes numériques
data.describe(include=[np.number])
```

```python
# Afficher les statistiques des variables catégoricals
data.describe(include=[object])
```

```python
# Les colonnes numériques
data_numerique = data.select_dtypes(include=[np.number])
data_numerique.head()
# Deuxième méthode : data.iloc[:,:-2]
```

```python
# Calculons la différence entre le max et le min de chaque colonne du data_numerique on utilisant la méthode
# apply
print(f'La difference entre la valeur maximal et la valeur minimal de chaque colonne est :\n')
print(f'{data_numerique.apply(lambda x:x.max()-x.min(),axis=0)}')
```

```python
## Réindexer le dataset
#Multiindex : ["model_year","origin"]
cols = ["model_year","origin"]
data.set_index(cols,inplace=True)
data.tail()
```

```python
# Trier index par ordre croissant
data.sort_index(ascending=True,inplace=True)
data.head()
```

```python
# noms des voitures
names =list(data.name)
# Les dix premières voitures
names[:10]
```

```python
# Afficher les voitures de l'année 75
data.loc[75]
```

```python
# Séléctionne toutes les voitures avec modele = japan,year=76
display(data.loc[76,"japan"])
print('################')
print(f'Il y a {data.loc[76,"japan"].shape[0]} modèles')
print('################')
```

```python
# Afficher les colonnes mpg et name, pour les années 77 et 79 avec comme pays d'origine japan

cols=["mpg","name"]
years = [77,79]
country ="japan"
display(data.loc[(years,country),cols])

#Afficher le mpg de datsun 210
print(f'The mpg of datsun 210 is : {data.loc[(years,country),cols].loc[79].reset_index().loc[1,"mpg"]}')
```

```python
# Changer l'ordre des indexes
data = data.swaplevel().sort_index(ascending=True)
data
```

```python
# Réinitialiser l'index, sauvegarder le résultat
data=data.reset_index(drop=False)
data.head()
```

```python
# Convertir la colonne origin en "title" : Les noms commence par majiscule
data.origin = data.origin.str.title()
```

```python
data
```

```python
# Extraire les noms des fabricants
split = data.name.str.split(pat = " ", n = 1, expand = True)
```

```python
# Liste des fabricants
fabricants = list(split[0].unique())
print(f'Le nombre de fabricants : {len(fabricants)}')
print(f'Voici la liste :')
display(fabricants)
```

```python
# Calculons leurs fréquences :
split[0].value_counts()
```

```python
# Le nombre des fabricants qui contient le mot "ford" dans la colonne name du dataset data
# Méthode 1 : on utilisant la fonction lambda
data[data.name.apply(lambda x:'ford' in str(x))]['name'].value_counts().sum()
```

```python
# Le nombre des fabricants qui contient le mot "ford" dans la colonne name du dataset data
# Méthode 2 : on utilisant la fonction contains
data.name.str.contains("ford").sum()
```
